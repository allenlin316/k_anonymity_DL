```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('adult.csv')

# Filter the DataFrame for income <= 50K
income_less_equal_50K = df[df['income'] == '<=50K']

# Filter the DataFrame for income > 50K
income_greater_than_50K = df[df['income'] == '>50K']

# Sample 75 rows from each filtered DataFrame
sampled_less_equal_50K = income_less_equal_50K.sample(n=150, random_state=1)
sampled_greater_than_50K = income_greater_than_50K.sample(n=150, random_state=1)

# Combine the sampled DataFrames
sampled_df = pd.concat([sampled_less_equal_50K, sampled_greater_than_50K])

# If you want to save the sampled data to a new CSV file
sampled_df.to_csv('sampled_income.csv', index=False)

# Display the sampled DataFrame
sampled_df.head()
```

```python
df = pd.read_csv("sampled_income.csv")
cnt_less = 0
cnt_more = 0
for income in df['income']:
    if income == "<=50K":
        cnt_less += 1
    else:
        cnt_more += 1
```
